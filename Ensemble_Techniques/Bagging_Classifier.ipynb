#**Multiple Heterogenous parellel Models are Built**

These are called heterogenous samples

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
x,y = make_classification(n_samples=1000,n_features=20,n_classes=2,random_state=1)
```

```python
x
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
lr_clf = LogisticRegression()
dt_clf = DecisionTreeClassifier()
svm_clf = SVC()
gnb_clf = GaussianNB()
```

#**Ensembling:**

```python
from sklearn.ensemble import VotingClassifier
```

```python
ensemble_clf = VotingClassifier(estimators=[('logistic regression',lr_clf),('decision_tree_clf',dt_clf),('support_vector_clf',svm_clf),('gaussian',gnb_clf)],voting = "hard")
```

```python
ensemble_clf
```

```python
ensemble_clf.fit(X_train,y_train)
```

```python
y_pred = ensemble_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
