```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
```

```python
x, y = make_classification(n_samples=1000, n_features=20, n_informative=15, n_redundant=5, random_state=1)
```

```python
x
```

```python
y
```

```python
len(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

#**Evaluation Matrix:**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve , auc
```

```python
y_pred_prob = model.predict_proba(x_test)[:,1]
```

```python
y_pred_prob
```

# **Caclulating ROC curve:**

```python
FPR, TPR, threshold = roc_curve(y_test, y_pred_prob)
```

```python
FPR
```

```python
TPR
```

```python
threshold
```

#**Calculating AUC score (Area under Curve)**

```python
roc_auc = auc(FPR, TPR)
```

```python
roc_auc
```

```python
plt.figure(figsize=(10,7))
plt.plot(FPR, TPR, color='darkorange',linewidth=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```

### **By default Prediction class is decided with CutOff = 0.5**

###**Sensitivity Specificity,accuracy TradeOff**

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score
```

```python
thresholds = np.linspace(0,1,100)
```

```python
y_pred_prob.shape
```

```python
Precisions = []
recalls = []
accuracies = []
```

```python
for threshold in thresholds:
  y_pred_threshold = (y_pred_prob >= threshold).astype(int)
  precision = precision_score(y_test, y_pred_threshold)
  recall = recall_score(y_test, y_pred_threshold)
  accuracy = accuracy_score(y_test, y_pred_threshold)
  Precisions.append(precision)
  recalls.append(recall)
```

```python
y_pred_threshold.shape
```

```python
plt.figure(figsize=(10,6))
plt.plot(thresholds, Precisions, label='Precision')
plt.plot(thresholds, recalls, label='Recall')
plt.plot(thresholds, accuracies, label='Accuracy')
plt.xlabel('Threshold Probability')
plt.ylabel('Scores')
plt.title('Precision, Recall and Accuracy Tradeoff')
plt.legend()
plt.grid(True)
plt.show()
```

```python
new_pred_levels = np.where(y_pred_prob>0.4,1,0)
```

```python
new_pred_levels
```

```python
confusion_matrix(y_test, new_pred_levels)
```

```python
accuracy_score(y_test, new_pred_levels)
```

```python
print(classification_report(y_test, new_pred_levels))
```
