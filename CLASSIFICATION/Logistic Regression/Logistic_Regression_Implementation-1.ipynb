# **Binary Classification:**

### **For Binary Classification we need only 2 classes. So we will filter out class-2**

```python
df = df[df['target']!=2]
```

```python
df
```

```python
df.target.unique()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

### **Train_Test_Split:**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1)
```

```python
x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

### **Model Training:**

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

```python
classifier
```

```python
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred
```

```python
classifier.predict_proba(x_test)
```

## **Evaluation Matrix:**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve , auc
```

### **Predicted Probabilities of CLass-2**

```python
classifier.predict_proba(x_test)[:,1]
```

```python
y_pred_prob = classifier.predict_proba(x_test)[:,1]
```

```python
y_pred_prob
```

### **ROC Curve:**

Inputs: y_test , y_pred_proba. It will return TPR,FPR with different cutoff of probabilities

```python
roc_curve(y_test, y_pred_prob)
```

```python
FPR , TPR , thresholds = roc_curve(y_test, y_pred_prob)
```

```python
FPR
```

```python
TPR
```

```python
thresholds
```

### **Area Under Curve (AUC):**

```python
roc_auc = auc(FPR,TPR)
```

```python
roc_auc
```

```python
plt.figure(figsize=(10,10))
plt.plot(FPR,TPR,color = 'orange',label='Logistic Regression (area = %0.2f)'%roc_auc)
plt.plot([0,1],[0,1],color='darkblue',linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.legend(loc='lower right')
plt.show()
```
