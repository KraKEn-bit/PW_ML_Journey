# **SCALING:**

So what I'm going to do is: Standardisation

## **Standardization**:

For this you need to find the value of mean and std from the column and find its z-score = (xi-mean)/std

Only from training data you can find the mean and std value. And based on this you apply it on the test dat.


###**Here**:

- 'Fit' means finding the value of mean   and std from train data.

- 'Transform' means applying that mean and std on the test data.


As test data is unseen you can't learn the mean and std from it.

So,

- fit and transform can be applied on **Train**

- Only transform can be applied on **Test**

```python
sns.distplot(df['total_bill'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler
```

```python
scaler.fit(df[['total_bill']])
# this will calculate mean(sigma) and std(standard deviation)
```

```python
scaler.transform(df[['total_bill']])
# This will apply the sigma and std on that column
```

```python
scaler.fit_transform(df[['total_bill']])

# Apply on multiple columns!
```

```python
pd.DataFrame(scaler.fit_transform(df[['total_bill','tip']]), columns=['total_bill','tip'])
```

# **If new data comes:**

```python
scaler.transform([[0.54,0.44]])
```
