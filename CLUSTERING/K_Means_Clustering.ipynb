```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=1)
```

```python
X
```

#**In Unsupervised Learning You'll not use y (Target Variable)**

```python
#in usl you will not use y
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3) #changing the value of n_clusters to see the result
kmeans
```

```python
kmeans.fit(X_train)
```

```python
y_labels = kmeans.predict(X_train)
```

```python
y_labels
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1], c = y_labels)
```

#**How better this clustering model is?**

```python
from sklearn.metrics import silhouette_score
silhouette_score(X_train, kmeans.labels_) #close to 1, we have a good model
```

```python
kmeans.labels_
```

```python
kmeans.inertia_ #k=3, this is wcss distance
```

#**How to select k(centroid)?**
Ans: Elbow method

```python
wcss = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss #wcss is decreasing as k value increases
```

```python
#plotting the elbow curve

plt.plot(range(1, 10), wcss)
plt.xticks(range(1, 10))
plt.xlabel("No of students")
plt.ylabel("WCSS")
plt.show()
```

### Based on the plot above the optimal no of k/centroid should be 3
