# **Problem Statement: I want to predict weight based on height divide the data into X and y**

```python
X = data['Height']
y = data['Weight']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 1)
```

## **N.B: LinearRegression expects X_train to be 2-dimensional, even if it has only one feature. But y_train can be 1-dimensional.**

Heres whats going on:

- X_train should be of shape (n_samples, n_features)

- y_train should be of shape (n_samples,)

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train.values.reshape(-1,1),y_train)
```

```python
plt.scatter(X_train,y_train,color = 'blue',label = 'training data')
plt.plot(X_test,model.predict(X_test.values.reshape(-1,1)),color = 'red',label = 'regression line')
plt.xlabel("Height")
plt.ylabel("Weight")
plt.legend()
plt.show()
```

## **Model Evaluation:**

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test,model.predict(X_test.values.reshape(-1,1)))
```
