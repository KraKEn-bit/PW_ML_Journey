```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
np.random.seed(1)
```

```python
X = 2*np.random.rand(100,1)
X
```

```python
y = 4+3*X +1.5*X + np.random.randn(100,1)
y
```

# **train_test split:**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# **Performing Polynomial Regression:**

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly_train = poly_features.fit_transform(X_train)
X_poly_train
```

```python
from sklearn.linear_model import LinearRegression

poly_reg = LinearRegression()
poly_reg.fit(X_poly_train,y_train)
```

```python
poly_reg.intercept_, poly_reg.coef_
```

# **Making prediction:**

```python
y_poly_train = poly_reg.predict(X_poly_train)
y_poly_train
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mean_squared_error(y_train,y_poly_train), r2_score(y_train,y_poly_train)
```

```python
plt.scatter(X_train,y_train,color='red',label='Training data')
degree = 2
X_range = np.linspace(0,2,100).reshape(-1,1)
X_range_poly = poly_features.transform(X_range)
plt.plot(X_range,poly_reg.predict(X_range_poly),color='blue',label=f'Polynomial Regression (degree={degree})')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression')
plt.legend()
plt.show()
```
