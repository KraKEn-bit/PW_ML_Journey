```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples=1000, n_features=10, n_informative=5, n_classes=2, random_state=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

#**HyperParameter Tuning:**

###**Grid Search CV and Random Search CV:**

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

###**Grid Search CV:**

```python
params = {
    'penalty': ['l1', 'l2', 'elasticnet'],
    'C': [1, 2, 10, 20, 30, 40],
    'solver': ['saga'],
    'l1_ratio': [0, 0.5, 1]
}
```

```python
params
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
classifier
```

```python
clf = GridSearchCV(classifier, param_grid=params, cv=5,verbose = 2)
```

```python
clf
```

```python
clf.fit(x_train, y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
clf.best_estimator_.predict(x_test)
```

```python
model = LogisticRegression(C=1,penalty='l2')
```

```python
model
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
model.predict_proba(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

###**Randomized Search CV:**

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
r_clf = RandomizedSearchCV(classifier, param_distributions=params,n_iter=10, cv=5)
```

```python
r_clf
```

```python
r_clf.fit(x_train, y_train)
```

```python
r_clf.best_params_
```

```python
r_clf.best_score_
```
