# **Outlier Handling:**

## **We see 'Age' , 'Fare' are continuous columns that may have outliers that need to be removed**

```python
# Histplot for age
plt.subplot(1,2,1)
sns.histplot(df_copy_copy['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution')

# Boxplot for Age
plt.subplot(1,2,2)
sns.boxplot(y=df_copy_copy['Age'], color='lightgreen')
plt.title('Age Boxplot')

plt.tight_layout()
plt.show()
```

```python
# Histplot for age
plt.subplot(1,2,1)
sns.histplot(df_copy_copy['Fare'], bins=20, kde=True, color='skyblue')
plt.title('Fare Distribution')

# Boxplot for Age
plt.subplot(1,2,2)
sns.boxplot(y=df_copy_copy['Fare'], color='lightgreen')
plt.title('Fare Boxplot')

plt.tight_layout()
plt.show()
```

### **We see outliers are present in 'Age' and 'Fare' column!**

```python
Q1 = df_copy_copy['Age'].quantile(0.25)
Q3 = df_copy_copy['Age'].quantile(0.75)

IQR = Q3 - Q1

lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR
```

```python
df_filtered = df_copy_copy[(df_copy_copy['Age'] >= lower_fence) & (df_copy_copy['Age'] <= upper_fence)]
```

```python
# Histplot for age
plt.subplot(1,2,1)
sns.histplot(df_filtered['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution')

# Boxplot for Age
plt.subplot(1,2,2)
sns.boxplot(y=df_filtered['Age'], color='lightgreen')
plt.title('Age Boxplot')

plt.tight_layout()
plt.show()
```

```python
df_filtered['Fare']
```

```python
Q1_prime = df_filtered['Fare'].quantile(0.25)
Q3_prime = df_filtered['Fare'].quantile(0.75)

IQR_prime = Q3_prime - Q1_prime

lower_fence_prime = Q1_prime - 1.5 * IQR_prime
upper_fence_prime = Q3_prime + 1.5 * IQR_prime
```

```python
df_filtered = df_filtered[(df_filtered['Fare'] >= lower_fence_prime) & (df_filtered['Fare']<= upper_fence_prime)]
```

```python
# Histplot for age
plt.subplot(1,2,1)
sns.histplot(df_filtered['Fare'], bins=20, kde=True, color='skyblue')
plt.title('Fare Distribution')

# Boxplot for Age
plt.subplot(1,2,2)
sns.boxplot(y=df_filtered['Fare'], color='lightgreen')
plt.title('Fare Boxplot')

plt.tight_layout()
plt.show()
```

# **EDA (Exploratory Data Analysis)**

```python
df_filtered_copy['Sex_female'].value_counts()
```

```python
df_filtered_copy['Sex_male'].value_counts()
```

```python
female_survival = df_filtered_copy[df_filtered_copy['Sex_female'] == 1]['Survived'].mean() * 100

male_survival = df_filtered_copy[df_filtered_copy['Sex_male']==1]['Survived'].mean() * 100

print(f"Female survival rate: {female_survival:.2f}%")
print(f"Male survival rate: {male_survival:.2f}%")
```

# **We see that the percentage of Female survival is greater than the survival rate of male!**

```python
df_filtered_copy['Sex'] = df_filtered_copy['Sex_female'].apply(lambda x: 'female' if x==1 else 'male')

plt.figure(figsize=(8,5))
sns.histplot(data=df_filtered_copy, x='Sex', hue='Survived', multiple='dodge', shrink=0.8)
plt.title('Survival Count by Sex')
plt.show()
```

# **Here's an analysis of how many people of what age had survived during the Titanic Incident!**

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df_filtered_copy, x='Age', hue='Survived', bins=30, kde=False, multiple='stack')
plt.title('Survival Count by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=df_filtered_copy[df_filtered_copy['Survived']==1]['Age'], label='Survived', shade=True)
sns.kdeplot(data=df_filtered_copy[df_filtered_copy['Survived']==0]['Age'], label='Not Survived', shade=True)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```
