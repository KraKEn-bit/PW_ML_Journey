# **How much employees have been terminated for each positions and reasons?**

```python
df.columns
```

```python
df.PositionID
```

The plot above shows the distribution of married and unmarried employees based on their gender. `MarriedID` 0 likely represents unmarried employees, and `MarriedID` 1 represents married employees. Similarly, `GenderID` 0 and 1 represent different genders. It appears there are more unmarried employees in general, and within both married and unmarried categories, there is a distribution between the two gender IDs.

```python
terminated_employees = df[df['EmploymentStatus'] == 'Terminated']

termination_counts = terminated_employees.groupby(['Position', 'TermReason']).size().reset_index(name='TerminationCount')

display(termination_counts.sort_values(by='TerminationCount', ascending=False))
```
