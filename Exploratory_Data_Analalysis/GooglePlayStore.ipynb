# **Why we reset_index()?**
Ans:
1. After filtering / cleaning, the index becomes messy

Example:
If you drop some rows:

df_copy = df_copy[df_copy['Age'] > 30]


Your DataFrame will now have indexes like:

5, 7, 8, 12, 14, ...


Resetting the index makes it:

0, 1, 2, 3, 4, ...

 2. drop=True removes the old index

If you use drop=False, the old index becomes a new column.

With drop=True, you delete the old index permanently.

 3. inplace=True modifies the DataFrame directly

Without this, you’d have to write:

df_copy = df_copy.reset_index(drop=True)


With inplace=True, it updates df_copy itself.

```python
df_copy.reset_index(drop=True, inplace=True)
df_copy
```

```python
def size_process(item):
    item = str(item).strip()

    # Case: value like "12M"
    if item.endswith("M"):
        res = float(item.replace("M", "")) * 1024
        return res

    # Case: value like "500k"
    elif item.endswith("k"):
        res = float(item.replace("k", ""))
        return res

    # No valid format
    else:
        return np.nan
```

```python
size_process('19M')
```

```python
df_copy['size'] = df_copy['Size'].apply(size_process)
```

```python
def simplify_android(ver):
    if pd.isna(ver):
        return np.nan
    # Split by '-' if range exists
    ver = ver.split('-')[0].strip()
    # Keep only major.minor (ignore patch)
    ver_parts = ver.split('.')
    if len(ver_parts) >= 2:
        return float(f"{ver_parts[0]}.{ver_parts[1]}")
    else:
        return float(ver_parts[0])
```

```python
df_copy['Android_Ver_Simplified'] = df_copy['Android Ver'].apply(simplify_android)
```

```python
def simplify_version(ver):
    if pd.isna(ver):
        return np.nan
    ver = str(ver).strip().split('-')[0].strip()
    if ver == 'Varies with device':
        return np.nan
    ver_parts = ver.split('.')
    if len(ver_parts) >= 2:
        try:
            return float(f"{ver_parts[0]}.{ver_parts[1]}")
        except ValueError:
            return np.nan
    elif len(ver_parts) == 1:
        try:
            return float(ver_parts[0])
        except ValueError:
            return np.nan
    return np.nan

df_copy['Current_Ver_Simplified'] = df_copy['Current Ver'].apply(simplify_version)
display(df_copy.head())
```
