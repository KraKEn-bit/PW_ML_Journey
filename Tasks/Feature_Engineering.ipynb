1) What is a parameter?
A parameter is a configuration variable internal to the model that is learned from the training data. For example, in linear regression, the coefficients (slopes) and intercept are parameters.

2) What is correlation?
Correlation measures the strength and direction of a linear relationship between two variables. It ranges from -1 to +1.

3) What does negative correlation mean?
Negative correlation means that as one variable increases, the other decreases. For example, hours spent playing video games and exam scores might have a negative correlation.

5) How does loss value help in determining whether the model is good or not?
The loss value quantifies how well the model's predictions match the actual results. A lower loss indicates better performance; a high loss indicates the model is underperforming.

6) What are continuous and categorical variables?

Continuous variables: Numerical values that can take any value within a range (e.g., height, temperature).

Categorical variables: Values representing categories or labels (e.g., gender, color).

7) How do we handle categorical variables in Machine Learning? What are the common techniques?
Categorical variables are handled using:

Label Encoding (converts categories to numbers)

One-Hot Encoding (creates binary columns for each category)

10) What is a Test set?
A test set is a portion of the data used to evaluate the performance of a trained model. It helps verify generalization to new, unseen data.

17) What is causation? Explain difference between correlation and causation with an example.
Causation means one variable directly affects another.
Correlation only shows an association.
Example:

Correlation: Ice cream sales and drowning increase together (due to summer).

Causation: More exercise causes weight loss.

18) What is an Optimizer? What are different types of optimizers? Explain each with an example.
An optimizer updates the model's parameters to minimize the loss.
Types:

SGD (Stochastic Gradient Descent): Updates weights based on each data point.

Adam: Combines momentum and adaptive learning rate.

RMSprop: Adapts learning rate per parameter.

27) Explain data encoding?
Data encoding transforms categorical data into numerical format:

Label Encoding: Converts each label to a number.

One-Hot Encoding: Creates separate binary columns for each category.

Example:

python
Copy
Edit

encoded = encoder.fit_transform(df[['Category']])
